# Analysis

•	Charter schools performed better than District schools in all categories, which led to a higher overall passing percentage of 90.43%. 
•	School sizes of 2,000 or smaller maintained passing averages in both reading and math and yielded an overall passing percentage of approximately 90%. 
•	Schools that spent $630 or less per student were also able to maintain passing averages in both math and reading and held an overall passing percentage of 81% or better.

```python
import pandas as pd
```

```python
schools_complete = "../Resources/schools_complete.csv"
students_complete = "../Resources/students_complete.csv"
```

```python
schools_complete_df = pd.read_csv(schools_complete)
students_complete_df = pd.read_csv(students_complete)
```

```python
schools_complete_df.head()
```

```python
students_complete_df.head()
```

```python
school_merge_df = pd.merge(schools_complete_df, students_complete_df, on="school_name", how="left")
school_merge_df.head()
```

```python
school_count = len(schools_complete_df["school_name"].unique())
school_count
```

```python
student_count = len(school_merge_df["Student ID"].unique())
student_count
```

```python
budget_count = sum(school_merge_df["budget"].unique()) 
budget_count
```

```python
math_avg_df = school_merge_df["math_score"].mean()
math_avg_df
```

```python
reading_avg_df = school_merge_df["reading_score"].mean()
reading_avg_df
```

```python
passing_math = school_merge_df.loc[school_merge_df['math_score'] >= 70]
pass_math_percent = len(passing_math)/student_count*100
pass_math_percent
```

```python
passing_reading = school_merge_df.loc[school_merge_df['reading_score'] >= 70]
pass_read_percent = len(passing_reading)/student_count*100
pass_read_percent
```

```python
pass_math_reading_count = school_merge_df[(school_merge_df["math_score"] >= 70) & (school_merge_df["reading_score"] >= 70)
].count()["student_name"]
overall_pass_rate = pass_math_reading_count /  float(student_count) * 100
overall_pass_rate
```

```python
district_summary = pd.DataFrame({"Total Unique Schools": [school_count], "Total Students": [student_count], "Total Budget": [budget_count], "Avg Math Score": [math_avg_df], "Avg Reading Score": [reading_avg_df], "% Pass Math": [pass_math_percent],"% Pass Reading": [pass_read_percent], "% Overall Pass": [overall_pass_rate]})

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

```python
schools_by_type = schools_complete_df.set_index(["school_name"])["type"]
schools_by_type
```

```python
schools_grouped = school_merge_df.groupby(['school_name'])
students_by_school = schools_grouped.size()
students_by_school
```

```python
total_school_budget = schools_complete_df.groupby(["school_name"]).mean()["budget"]
total_school_budget
```

```python
per_school_budget = schools_complete_df.groupby(["school_name"]).mean()["budget"]
capita_per_school = per_school_budget / students_by_school
capita_per_school
```

```python
avg_math_per_school = school_merge_df.groupby(["school_name"]).mean()["math_score"]
avg_math_per_school
```

```python
avg_read_per_school = school_merge_df.groupby(["school_name"]).mean()["reading_score"]
avg_read_per_school
```

```python
#passing_math_per_school = school_merge_df.groupbyschool_merge_df.groupby(["school_name"]).loc()["math_score"] >= 70]
passing_math_per_school = school_merge_df[(school_merge_df["math_score"]>=70)]
pass_math_percent = passing_math_per_school.groupby(["school_name"]).count()["student_name"] / students_by_school * 100
pass_math_percent
```

```python
passing_read_per_school = school_merge_df[(school_merge_df["reading_score"]>=70)]
pass_read_percent = passing_read_per_school.groupby(["school_name"]).count()["student_name"] / students_by_school * 100
pass_read_percent
```

```python
passing_math_and_reading = school_merge_df[(school_merge_df["reading_score"] >= 70) & (school_merge_df["math_score"] >= 70)]
passing_math_and_reading_perc = passing_math_and_reading.groupby(["school_name"]).count()["student_name"] / students_by_school * 100
passing_math_and_reading_perc
```

```python
per_school_summary = pd.DataFrame({"School Type": schools_by_type, "Total Students": students_by_school,
                                  "Total School Budget": total_school_budget, "Per Student Budget": capita_per_school,
                                  "Math Average": avg_math_per_school, "Reading Average": avg_read_per_school,
                                  "% Pass Math": pass_math_percent, "% Pass Reading":pass_read_percent,
                                  "% Overall Passing": passing_math_and_reading_perc})
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary                                  
```

```python
per_school_summary_noformat = per_school_summary
per_school_summary_noformat.head()
```

```python
school_spending = per_school_summary
school_spending
```

```python
school_spending["Spending Ranges (Per Student)"] = pd.cut(per_school_summary_noformat["Per Student Budget"], spending_bins, labels=spending_labels)
school_spending
```

```python
spending_math_scores = school_spending.groupby(["Spending Ranges (Per Student)"]).mean()["Math Average"]
spending_reading_scores = school_spending.groupby(["Spending Ranges (Per Student)"]).mean()["Reading Average"]
spending_passing_math = school_spending.groupby(["Spending Ranges (Per Student)"]).mean()["% Pass Math"]
spending_passing_reading = school_spending.groupby(["Spending Ranges (Per Student)"]).mean()["% Pass Reading"]
overall_passing_spending = school_spending.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]
```

```python
spending_summary= pd.DataFrame({"Math Average": spending_math_scores, "Reading Average": spending_reading_scores, "Pass Math": spending_passing_math, "% Pass Reading": spending_passing_reading, "% Overall Passing": overall_passing_spending })
spending_summary
```
